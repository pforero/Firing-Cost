# Firing Cost Structure

Firing Cost Function creates a grid with the adjustment costs paid by the firm for firing a worker for each level of tenure.

### Input Variables:

$G$ Max number of tenure workers  
$\phi$ Percentage of Wage Paid as Firing Cost  
$S=\left\{ \begin{array}{l l}
1 & \text{Seniority}\\
0 & \text{No Seniority}
\end{array} \right.$ Increase of Firing Cost per year of tenure  
$W$ Wage Structure

### Output Variables:

$\mathbb{FC}$ Grid with the Firing Cost of each Tenure Worker

## Function:

```julia
function FiringCost(G::Int,ϕ::Float64,S::Bool,W::Array{Float64,1})
    
    """Create a grid with the adjustment costs 
    paid by the firm for firing a worker for 
    each level of tenure.

    Parameters
    ----------
    G: Max Number of Tenure Workers
    ϕ: Firing Cost as Share of Wage 
    S: Seniority Dummy
    W: Wage Structure
    
    Output
    ----------
    FC: Grid with the Firing Cost of each Tenure Worker
    """
    
    TGrid=[-1:G-2]*S+1;
    
    FC=TGrid.*(ϕ*W);
    
    return FC
    
end
```

## Example

### Parameters:

```julia
G=10;
ϕ=0.1;
S=true;
W=ones(G);
```

```julia
FCStructure=FiringCost(G,ϕ,S,W);
```

```julia
using Gadfly
using Color

plot(x=[0:G-1]+1/2,y=FCStructure,
color=FCStructure,
Scale.ContinuousColorScale(Scale.lab_gradient(color(RGB(202/255,215/255,237/255)),color(RGB(37/255,131/255,192/255)),color(RGB(0/255,74/255,124/255)))),
Guide.ylabel("Firing Cost"),
Guide.xlabel("Tenure"),
Guide.title("Firing Cost Structure"),
Guide.xticks(ticks=[0:G]),
Geom.bar(position=:dodge))
```
