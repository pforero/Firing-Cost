# Feasible Labour Portfolio Choices (Parallel)

#### Add Processors:

```julia
addprocs(3)
```

### Input Variables:

$\textit{f}$ Function used to Evaluate Possible Portfolios  
$\mathbb{L}$ Labour Portfolios   

### Output Variables:

$\mathbf{R}$ Array of Indexes of Today Portfolio by the Number of Possible Portfolios  
$\mathbf{C}$ List with all possible Portfolio indexes for each Portfolio

## Function:

```julia
function FeasibleQnRP(f::Function,L::Array{Any,1})
    
    """Returns the Array with the 
    Index Value of all Feasible future
    Portfolios by Quits and Retirements.
    
    Parallel Version

    Parameters
    ----------
    f : Function used to create Feasible Portfolios
    L : Labour Portfolios
    
    Parameters
    ----------
    R : Array of Indexes of Today Portfolio by the Number of Possible Portfolios
    C : Array of Indexes of Possible Portfolios for tomorrow due to QnR
    """
    
    C=Array(Array{Int64,1},length(L));
    
    H=pmap(f,L);
    
    for i=1:length(L)
        
        C[i]=findin(L,H[i]);
        
    end

    R=copy(C);
    
    for i=1:length(L)
        
        fill!(R[i],i);
        
    end

    return R,C
    
end
```

## Growing Size

$\mathbb{L}(G)$ 

```julia
import LabourDP

N=3;
M_D=9;
```

## Time

Max Size and Sample Size

```julia
Ms=5;
Ss=1;
```

### Test

```julia
# Warm up Run
G=3;
fPortfolio=LabourDP.fPort(G);
Labour=LabourDP.LabourPortfolio(G,N,M_D,fPortfolio);
FeasibleQnRP(fPortfolio,Labour); #

x=[1:Ms];
t=[1.0:Ms];
b=[1.0:Ss];

for i=1:Ms
    
    G=i+1;
    fPortfolio=LabourDP.fPort(G);
    Labour=LabourDP.LabourPortfolio(G,N,M_D,fPortfolio);
    
    x[i]=length(Labour);
    
    for j=1:Ss
    
        a=@elapsed FeasibleQnRP(fPortfolio,Labour)
        b[j]=a;
        
    end
    
    t[i]=minimum(b);
    
end
```

### Plot Result

```julia
using Gadfly

plot(x=x, y=t,Guide.xlabel("Size L"),Guide.ylabel("Seconds"),Guide.title("Time"), Geom.line, Geom.point)
```
